### Data Visualization

```python
# Separate numeric and categorical columns
numeric_columns = df.select_dtypes(exclude=['object'])
categorical_columns = df.select_dtypes(include=['object'])
```

```python
# Plot numeric variable distributions with mean and median
for col in numeric_columns.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=10, color='blue', label='Data')
    
    # Calculate mean and median
    mean = df[col].mean()
    median = df[col].median()
    
    # Add vertical lines for mean and median
    plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
    plt.axvline(median, color='green', linestyle='-', label=f'Median: {median:.2f}')
    
    # Add plot details
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Plot categorical variable distributions
for col in categorical_columns.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=df[col], palette='Set2')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.grid(True, axis='y')
    plt.show()
```

```python
# Select one numeric column and one categorical column
numeric_col = 'Family_Income'
categorical_col = 'Gender'

# Plot distribution and mean
plt.figure(figsize=(8, 6))

# Boxplot for distribution
sns.boxplot(data=df, x=categorical_col, y=numeric_col, palette='Set2', showmeans=True, meanprops={"marker": "o", "markerfacecolor": "red", "markeredgecolor": "black"})

# Add title and labels
plt.title(f'Distribution and Mean of {numeric_col} by {categorical_col}')
plt.xlabel(categorical_col)
plt.ylabel(numeric_col)
plt.grid(True, axis='y')
plt.show()
```

```python
# Select one numeric column and one categorical column
numeric_col = 'Family_Income'
categorical_col = 'Region'

sns.barplot(data=df, x=categorical_col, y=numeric_col, ci='sd', palette='Set2')
plt.title(f'Mean {numeric_col} by {categorical_col} with Error Bars')
plt.xlabel(categorical_col)
plt.ylabel(f'Mean {numeric_col}')
plt.grid(True, axis='y')
plt.show()
```
